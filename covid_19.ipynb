```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
```

****load a dataset from kaggle:"https://www.kaggle.com/datasets/georgesaavedra/covid19-dataset?utm_source=chatgpt.com"****

```python
pd.set_option("display.max_column",None)
data=pd.read_csv("owid-covid-data.csv")
data.head()
```

```python
data.shape
```

```python
data.info()
```

**split a location by country**

```python
data["location"].value_counts().sort_index()
```

****make a new dataset by selected country from the main dataset****

```python
country=["India","Sri Lanka","Thailand","South Korea"]
covid_data=data[data["location"].isin(country)]
covid_data.shape
```

****perform a timeserise****

```python
covid_data.loc[:,"date"]=pd.to_datetime(covid_data["date"])
```

```python
pd.set_option("display.max_rows",None)
covid_data.isna().sum()
```

****handle a missing value****

```python
covid_data.loc[:,"new_cases"]=covid_data["new_cases"].fillna(0)
covid_data.loc[:,"new_deaths"]=covid_data["new_deaths"].fillna(0)
```

****make a day datset****

```python
daily_case=covid_data.groupby(["date","location"])["new_cases"].sum()
daily_case.shape
```

****perform a weekly cases***

```python
weekly_cases = covid_data.groupby(
    ["location",
     pd.Grouper(key="date",freq="W")]
)["new_cases"].sum()

```

```python
covid_data.loc[:,"rolling_avg"] = covid_data.groupby(
    "location"
)["new_cases"].transform(
    lambda x: x.rolling(7).mean()
)
```

****plot about "date and rolling_avg"****

```python
plt.figure(figsize=(12,6))
for country in country:
    temp = covid_data[covid_data['location']==country]
    plt.plot(
        temp['date'],
        temp["rolling_avg"],
        label=country
    )
plt.legend()
plt.title('7-Day Rolling Average Cases')
plt.xlabel('Date')
plt.ylabel('Cases')
plt.show()
```

****plot a peak cases and plot****

```python
peak_cases = covid_data.groupby(
    'location'
)['new_cases'].max()

print(peak_cases)
```

```python
peak_cases.plot(kind='bar')
plt.title('Peak Daily Cases')
plt.ylabel('Cases')
plt.show()
```

****make a new coumn form old dataset by grouping location & new_cases****

```python
covid_data['R_ratio'] = covid_data.groupby('location')['new_cases'].pct_change()
print(covid_data[['location','date','R_ratio']].head())
```
